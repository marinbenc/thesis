```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# prewitt filter
kernel = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
img = cv.imread('images/windowing-example.png', 0)
img = img[:, 620:]

conv = cv.filter2D(img, -1, kernel)
# stack images side-by-side
img = np.hstack((img, conv))
plt.imshow(img, cmap='gray')
# save img
# cv.imwrite('images/prewitt-example.png', img)
```
